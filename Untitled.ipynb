## The objective of this data science project is to predict miles per gallon (mpg) for a given car.

Before seeing the data let us answer 3 questions:

1. Is it a supervised, unsupervised, or Reinforcement Learning?
 This is a supervised learning. We know that we have to predict miles per gallon (mpg).
 
2. Is it a classification task, a regression task, or something else?
 We have to predict Miles per gallon. mpg is a measure of fuel economy of a car and it is a numerical value. Hence this is a regression task.
 
3. Should we use batch learning or online learning techniques?
 As there is no continuous flow of data coming into the system, so batch learning is enough.

We have 7 numeric (3 continuous and 4 discrete) and 2 object columns. Don't you think that 'horsepower' should be a numeric value and not an object? Then, let us change the datatype of horsepower from object to numeric.

```python
df['horsepower'] = df['horsepower'].astype('float64')
```

Let us check how many rows have '?' in 'horsepower' columns.

```python
df.loc[df['horsepower'] == '?']
```

#### Since the number of rows with ? is minimal we can eliminate them

```python
df = df.loc[df['horsepower']!='?']
```

```python
#Now changing the data type to float
df['horsepower'] = df.horsepower.astype('float64')
```

#### Since car name should have no impact on our model we drop those values and convert our categorical origin  and cylinder values to actual values after segregrating them

```python
#drop car name
df = df.drop('car name', axis=1)
```

```python
#replacing categorical origins with actual values
df['origin'] = df['origin'].replace({1: 'america', 2: 'europe', 3: 'asia'})
df.head()
```

```python
#converting origin data type to string
df["origin"] = df["origin"].astype(str)
```

```python
#converting cylinder data type to string
df["cylinders"] = df["cylinders"].astype(str)
```

```python
# One Hot Encoding - 1
df = pd.get_dummies(df,drop_first=False)
```

```python
# I decide to use model year, so i will apply one hot encoding
df["model year"] = df["model year"].astype(str)
```

```python
# One Hot Encoding - 2
df = pd.get_dummies(df,drop_first=False)
```

```python
df.info()
```

#### Train Test Split

```python
y = df["mpg"]
x = df.drop(["mpg"],axis=1)
```

```python
#Importing Libraries
from sklearn.preprocessing import RobustScaler,StandardScaler
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import mean_squared_error
from sklearn.base import clone
```

```python
# Standardization
x = StandardScaler().fit_transform(x)
```

```python
# Creating Train and Test Datasets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.90, random_state=42)
```

We use linear regression to create our model

```python
regressor = LinearRegression()
```

```python
regressor.fit(x_train,y_train)
```

```python
y_predicted = regressor.predict(x_test)
```

```python
mse = mean_squared_error(y_test,y_predicted)
print("MSE",mse)
```

#### We use Regularization Technics since it discourages learning a more complex or flexible model, so as to avoid the risk of overfitting.

#### We start off with Ridge and then proceed to Lasso Regularization

```python
ridge = Ridge(random_state=42, max_iter=10000)
alphas = np.logspace(-4,-0.5,30)
tuned_parameters = dict(alpha=alphas)
```

```python
clf = GridSearchCV(ridge,tuned_parameters,cv=6,scoring="neg_mean_squared_error", refit=True)
clf.fit(x_train,y_train)
scores = clf.cv_results_["mean_test_score"]
scores_std = clf.cv_results_["std_test_score"]
```

```python
print("Ridge Coef:",clf.best_estimator_.coef_)
ridge = clf.best_estimator_
print("Ridge Best Estimator:",ridge)
```

```python
y_pred_ridge = clf.predict(x_test)
mse_ridge = mean_squared_error(y_test,y_pred_ridge)
print("Ridge MSE:",mse_ridge)
```

```python
lasso = Lasso(random_state=42, max_iter=10000)
alphas = np.logspace(-4,-0.5,30)
tuned_parameters = dict(alpha=alphas)
```

```python
clf = GridSearchCV(lasso,tuned_parameters,cv=5,scoring="neg_mean_squared_error", refit=True)
clf.fit(x_train,y_train)
scores = clf.cv_results_["mean_test_score"]
scores_std = clf.cv_results_["std_test_score"]
```

```python
print("Lasso Coef:",clf.best_estimator_.coef_)
lasso = clf.best_estimator_
print("Lasso Best Estimator:",lasso)
print("Put Zero for redundat features:")
```

```python
y_pred_lasso = clf.predict(x_test)
mse_lasso = mean_squared_error(y_test,y_pred_lasso)
print("Lasso MSE:",mse_lasso)
```
